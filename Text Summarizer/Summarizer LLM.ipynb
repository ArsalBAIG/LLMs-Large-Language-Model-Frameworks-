```python
import numpy as np
import pandas as pd
import transformers
import torch

from transformers import AutoTokenizer
from transformers import pipeline
pipe = pipeline("summarization", model="facebook/bart-large-cnn")
```

# Loading Model & Tokenizer

```python
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
pipeline = transformers.pipeline(
    "summarization", model="facebook/bart-large-cnn", tokenizer=tokenizer, torch_dtype= torch.float16, min_length=60, max_length=100
)
```

```python
# Summarization Function.
def summarize_text(text):
    summary = pipeline(text, min_length=30, max_length=80)
    return summary[0]['summary_text']
```

# Test Examples

```python
example_text1 = """\n    Artificial intelligence (AI) is intelligence demonstrated by machines, \n    unlike the natural intelligence displayed by humans and animals. Leading AI textbooks \n    define the field as the study of \"intelligent agents\": any device that perceives its \n    environment and takes actions that maximize its chance of successfully achieving \n    its goals. Colloquially, the term \"artificial intelligence\" is often used to describe \n    machines that mimic \"cognitive\" functions that humans associate with the human mind, \n    such as \"learning\" and \"problem-solving\".\n    """

summarized_content = summarize_text(example_text1)
print(example_text1)
print(summarized_content)
```

```python
example_text2 = '''Electricity is a powerful form of energy that runs almost every modern device. we use It moves through wires to light our homes,
power machines, and keep technology working smoothly.
Without electricity, daily life would become difficult,
as many essential systems depend on it. Its importance continues to grow as the world becomes more advanced and connected.\n'''

summaried_content = summarize_text(example_text2)
print(example_text2)
print(summaried_content)
```

# Making & Saving the Model & Tokenizer

```python
import os
save_directory = "./summarization_model"

if not os.path.exists(save_directory):
    os.makedirs(save_directory)

print(f"Saving model and tokenizer to: {save_directory}")
```

```python
# Saving Model & Tokenizer.
tokenizer.save_pretrained(save_directory)
pipeline.model.save_pretrained(save_directory)
```
